```python
#Initially, null values were simply removed. However, employing the method of using the means of the features to 
#replace null values yielded a higher r^2 score - and was thus kept. 
mean_GDP=df['Log GDP per capita'].mean()
df['Log GDP per capita'].fillna(value=mean_GDP, inplace=True)
mean_SS=df['Social support'].mean()
df['Social support'].fillna(value=mean_SS, inplace=True)
mean_demQ=df['Democratic Quality'].mean()
df['Democratic Quality'].fillna(value=mean_demQ, inplace=True)
mean_delQ=df['Delivery Quality'].mean()
df['Delivery Quality'].fillna(value=mean_delQ, inplace=True)
mean_lifeExp=df['Healthy life expectancy at birth'].mean()
df['Healthy life expectancy at birth'].fillna(value=mean_lifeExp, inplace=True)
```

```python
nan_count = np.sum(df[features].isnull(), axis = 0)
nan_count_label = np.sum(df['Healthy life expectancy at birth'].isnull(), axis = 0)

print(nan_count)
print(nan_count_label)
```

```python
features = ['Life Ladder', 'Log GDP per capita', 'Social support', 'Delivery Quality', 'Democratic Quality']
X = df[features]
X.head()
y = df['Healthy life expectancy at birth']
print(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.10, random_state=1234)
```

```python
# Create the  LinearRegression model object 
model = LinearRegression()

# Fit the model to the training data 
model.fit(X_train, y_train)

#  Make predictions on the test data 
prediction = model.predict(X_test)
```

```python
print('Model Summary:\n')

# Print intercept (alpha)
print('Intercept:')
print('alpha = ' , model.intercept_)

# Print weights
print('\nWeights:')
i = 0
for w in model.coef_:
    print('w_',i+1,'= ', w, ' [ weight of ', features[i],']')
    i += 1
```

```python
# Print mean squared error
print('\nModel Performance\n\nRMSE =   %.2f'
      % np.sqrt(mean_squared_error(y_test, prediction)))
# The coefficient of determination
print(' R^2 =   %.2f'
      % r2_score(y_test, prediction))
```

The RMSE value of 3.85 and R^2 value of 0.76 indicates that the regression model fits the data fairly well.
